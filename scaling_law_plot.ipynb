```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def parse(result):
    result_dict = dict()
    for line in result.split('\n'):
        tmp = line.split(' ')
        key = tmp[1][:-1]
        value = float(tmp[2])
        result_dict[key] = value
    return result_dict

def proof_scaling_law_plot(sizes, times):
    plt.plot(sizes, times)
    plt.xlabel('number of parameters')
    plt.ylabel('proof generation time')
    plt.title('Scaling law for proof generation (EZKL)')
    plt.show()

def verify_scaling_law_plot(sizes, times):
    plt.plot(sizes, times)
    plt.xlabel('number of parameters')
    plt.ylabel('verification time')
    plt.title('Scaling law for verification (EZKL)')
    plt.show()
```
